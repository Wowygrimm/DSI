```python
import cv2
import numpy as np
import pandas as pd
import glob
import os
```

```python
path = 'C:\dlib-19.17.0\python_examples\Eye_only\\train'
```

```python
for f in glob.glob(os.path.join(path, "*.jpg")):
    print(f)
```

```python
Eye_matrix = pd.DataFrame(columns = ['i0', 'i1', 'i2', 'i3', 'i4', 
                                     'i5', 'i6', 'i7', 'i8', 'i9',  
                                     'i10', 'i11', 'i12', 'i13', 'i14', 'Position'])
```

```python
for f in glob.glob(os.path.join(path, "*.jpg")):
    
    filename = f.replace(path+"\\","")
    filename = filename.replace("_eye_only.jpg","")
    
    img = cv2.imread(f, 0)
    height, width = img.shape

    step_h = height//3
    step_w = width//5

    level = []
    light_intensity = 0
    for i in range(15):
        tile = img[step_h*(i%3):step_h*(i%3 + 1),step_w*(i%5):step_w*(i%5 + 1)]
        tile_value = tile.mean(axis = 0).mean(axis = 0)
        light_intensity = light_intensity + tile_value
        level.append(tile_value)

    level_normed = [x / light_intensity for x in level]
    level_normed.append(filename)
    Eye_matrix = Eye_matrix.append({'i0':level_normed[0], 'i1':level_normed[1], 'i2':level_normed[2], 'i3':level_normed[3], 'i4':level_normed[4], 
                                    'i5':level_normed[5], 'i6':level_normed[6], 'i7':level_normed[7], 'i8':level_normed[8], 'i9':level_normed[9], 
                                    'i10':level_normed[10], 'i11':level_normed[11], 'i12':level_normed[12], 'i13':level_normed[13], 'i14':level_normed[14], 'Position':level_normed[15]}, ignore_index=True)
```
